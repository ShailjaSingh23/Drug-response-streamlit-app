**Task 4: SVM Implementation**

```python
from sklearn.svm import SVC
```

```python
# Create model
svm_model = SVC()

# Train model
svm_model.fit(X_train, Y_train)
```

```python
y_pred = svm_model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy = accuracy_score(Y_test, y_pred)
precision = precision_score(Y_test, y_pred)
recall = recall_score(Y_test, y_pred)
f1 = f1_score(Y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

Model is performing well as it is giving 82% accuracy with 86% of precision, 80% of recall and 83% of f1 score.

**Task 5: Visualization of SVM Results**

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

The confusion matrix shows that the model correctly classified 82 out of 100 test samples, achieving 82% accuracy, with 7 false positives and 11 false negatives.

**Task 6: Parameter Tuning and Optimization:** Experiment with different SVM hyperparameters (e.g., kernel type, regularization parameter) to optimize performance.

```python
# Linear kernel
svm_linear = SVC(kernel='linear', C=1)
svm_linear.fit(X_train, Y_train)
y_pred_linear = svm_linear.predict(X_test)
print("Linear Accuracy:", accuracy_score(Y_test, y_pred_linear))
```

```python
# RBF kernel
svm_rbf = SVC(kernel='rbf', C=1, gamma='scale')
svm_rbf.fit(X_train, Y_train)
y_pred_rbf = svm_rbf.predict(X_test)
print("RBF Accuracy:", accuracy_score(Y_test, y_pred_rbf))
```

```python
# Polynomial kernel
svm_poly = SVC(kernel='poly', C=1, degree=3)
svm_poly.fit(X_train, Y_train)
y_pred_poly = svm_poly.predict(X_test)
print("Polynomial Accuracy:", accuracy_score(Y_test, y_pred_poly))
```

testing model with different C value

```python
for c in [0.1, 1, 10, 100]:
    model = SVC(kernel='rbf', C=c)
    model.fit(X_train, Y_train)
    y_pred = model.predict(X_test)
    print("C =", c, "Accuracy:", accuracy_score(Y_test, y_pred))
```

```python
for c in [0.1, 1, 10, 100]:
    model = SVC(kernel='poly', C=c)
    model.fit(X_train, Y_train)
    y_pred = model.predict(X_test)
    print("C =", c, "Accuracy:", accuracy_score(Y_test, y_pred))
```

```python
for c in [0.1, 1, 10, 100]:
    model = SVC(kernel='linear', C=c)
    model.fit(X_train, Y_train)
    y_pred = model.predict(X_test)
    print("C =", c, "Accuracy:", accuracy_score(Y_test, y_pred))
```

**Task 7: Comparison and Analysis:** Compare SVM performance with various kernels (e.g., linear, polynomial, radial basis function).

```python
linear_acc = 0.78
poly_acc = 0.69
rbf_acc = 0.82

print("Linear Kernel Accuracy:", linear_acc)
print("Polynomial Kernel Accuracy:", poly_acc)
print("RBF Kernel Accuracy:", rbf_acc)

best = max(linear_acc, poly_acc, rbf_acc)

if best == linear_acc:
    print("Best Kernel: Linear")
elif best == poly_acc:
    print("Best Kernel: Polynomial")
else:
    print("Best Kernel: RBF")
```

**SVM Strengths & Weaknesses (Based on EDA)**

Strengths: Works well for high-dimensional data, Effective for small to medium datasets and Robust to overfitting (with proper C).


Weaknesses: Not ideal for very large datasets (slow training) and Sensitive to hyperparameters (C, gamma).
